```python
from matplotlib import pyplot as plt
plt.style.use("seaborn-whitegrid")
import numpy as np
print("step 1")
```

```python
fig =plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig =plt.figure()
ax=plt.axes()
x=np.linspace(0,20,1000)
#print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
ax.set_title('Simplest Plot')
ax.set_xlabel('Wavelength')
ax.set_ylabel('amplitude');
ax.legend()
plt.show()
```

```python
fig =plt.figure()
ax=plt.axes()
x=np.linspace(0,20,1000)
#print(x)
ax.plot(x,np.sin(x),label='sin',color='g',linestyle='dashed');
ax.plot(x,np.cos(x),label='cos',color='yellow',linestyle='dotted');
ax.plot(x,np.sin(x+1),label='sinn',linestyle='dashdot');
ax.set_title('Simplest Plot')
ax.set_xlabel('Wavelength')
ax.set_ylabel('amplitude');
ax.legend()
plt.show()
```
