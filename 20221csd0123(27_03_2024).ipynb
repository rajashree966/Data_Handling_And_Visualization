```python
import pandas as pd
df=pd.read_csv('/content/dataset/train.csv')
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df.tail()
```

```python
age_mean_value=df['Age'].mean()
df.head()
```

```python
df.tail()
```

```python
df["Age"]=df['Age'].fillna(age_mean_value)
df.head()
```

```python
df.tail()
```

```python
df.drop("Cabin",axis=1,inplace=True)
df.head()
```

```python
df.tail()
```

```python
filtered_age=df[df.Age>40]
filtered_age
```

```python
sorted_passengers=df.sort_values("Name",ascending=True,kind='heapsort')
sorted_passengers.head(10)
```

```python
merged_df=pd.merge(df.head(2),df.tail(2),how='outer',indicator='True')
merged_df
```

```python
group_df=df.groupby('Sex')
for key,item in group_df:
  print(group_df.get_group(key),"\n\n")
```
