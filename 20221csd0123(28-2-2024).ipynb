```python
from matplotlib import pyplot as plt
plt.style.use("seaborn-whitegrid")
import numpy as np
print("step 1")
```

```python
fig =plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig =plt.figure()
ax=plt.axes()
x=np.linspace(0,20,1000)
#print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
ax.set_title('Simplest Plot')
ax.set_xlabel('Wavelength')
ax.set_ylabel('amplitude');
plt.show()
```

```python
fig =plt.figure()
ax=plt.axes()
x=np.linspace(0,20,1000)
#print(x)
ax.plot(x,np.sin(x),label='sin',color='g',linestyle='dashed');
ax.plot(x,np.cos(x),label='cos',color='yellow',linestyle='dotted');
ax.plot(x,np.sin(x+1),label='sinn',linestyle='dashdot');
ax.set_title('Simplest Plot')
ax.set_xlabel('Wavelength')
ax.set_ylabel('amplitude');
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.set_xlim(-5,15)
ax.set_ylim(-3,3);
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
x=np.linspace(0,10,1000)
axs[0,0].plot(x,np.sin(x))
axs[0,1].plot(x,np.cos(x),color='maroon')
axs[1,0].plot(x,np.sin(x+2),color='blue')
axs[1,1].plot(x,np.sin(x+4),color='green');
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Lime':2}
names=list(data.keys())
values=list(data.values())
fig=plt.figure()
ax=plt.axes()
ax.bar(names,values)
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Lime':2}
names=list(data.keys())
values=list(data.values())
fig=plt.figure()
ax=plt.axes()
ax.barh(names,values)
```

```python
d={'Apple':10,'Orange':15,'Banana':13,'Pear':9,'Pomogranate':11,'Lemon':5,'Lime':20}
mykeys=list(d.keys())
mykeys.sort()
data={i:d[i] for i in mykeys}
names=list(data.keys())
values=list(data.values())
fig=plt.figure()
ax=plt.axes()
ax.barh(names,values)
ax.set_xlabel("Quantity")
ax.set_ylabel("Fruits")
ax.set_title(" Fruits Quantities");
```

```python
labels=['G1','G2','G3','G4','G5']
men=[20,34,30,35,27]
women=[25,32,34,20,25]
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men,width,label='men')
rects2=ax.bar(x+width/2,women,width,label='women')
ax.set_ylabel("Scores")
ax.set_title("Scores by group and gender")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
ax.bar_label(rects1,padding=3)
ax.bar_label(rects2,padding=3)
ax.set_ylim(0,40)
```
